Tasks:
1. Load the dataset using Pandas.
2. Display first and last five rows.
3. Find shape, columns, and data types.
4. Generate descriptive statistics.


```python
import pandas as pd
df = pd.read_csv('/content/Student_Performance.csv')
#printing the First Five rows
print("The First Five Rows: ")
display(df.head())
print()
#printing the last Five Rows
print("The Last Five Rows: ")
display(df.tail())
print()
#Shape of the dataframe
print("Shape of the DataFrame:", df.shape)
print()
#Columns of dataframe
print("Columns of the DataFrame:", df.columns.tolist())
print()
#Data types and non-null counts
print("Data types and non-null counts: ")
df.info()
print()
#Descriptive statistics
print("Descriptive statistics: ")
display(df.describe(include='all'))
print()
```


Tasks:
1. Select score-related columns.
2. Filter students scoring above 70 in math.
3. Filter data based on gender.
4. Count number of students in each category.



```python
# 1. Select score-related columns
score_columns = df[['math_score', 'science_score', 'english_score', 'overall_score']]
print("\nScore-Related Columns (first 5 rows):")
display(score_columns.head())
print()
# 2. Filter students scoring above 70 in math
students_above_70_math = df[df['math_score'] > 70]
print("\nStudents with Math Score > 70 (first 5 rows):")
display(students_above_70_math.head())
print(f"Number of students with Math Score > 70: {len(students_above_70_math)}")
print()

# 3. Filter data based on gender (example: 'female')
female_students = df[df['gender'] == 'female']
print("\nFemale Students (first 5 rows):")
display(female_students.head())
print(f"Number of female students: {len(female_students)}")
print()

male_students = df[df['gender'] == 'male']
print("\nMale Students (first 5 rows):")
display(male_students.head())
print(f"Number of male students: {len(male_students)}")
print()

other_gender_students = df[df['gender'] == 'other']
print("\nOther Gender Students (first 5 rows):")
display(other_gender_students.head())
print(f"Number of other gender students: {len(other_gender_students)}")
print()
# 4. Count number of students in each category (example: 'gender' and 'school_type')
print("\nStudent count by Gender:")
display(df['gender'].value_counts())
print()

print("\nStudent count by School Type:")
display(df['school_type'].value_counts())
print()

print("\nStudent count by attendance_percentage:")
display(df['attendance_percentage'].value_counts())
print()
```

Task 3:Tasks:
1. Convert score columns into NumPy arrays.
2. Compute mean, median, standard deviation.
3. Find minimum and maximum scores


```python
import numpy as np

# Score columns to analyze
score_columns = ['math_score', 'science_score', 'english_score', 'overall_score']

# Convert score columns to NumPy arrays and perform calculations
for col in score_columns:
    # 1. Convert score column into NumPy array
    score_array = df[col].to_numpy()
    print(f"\n--- {col.replace('_', ' ').title()} Statistics ---")

    # 2. Compute mean, median, standard deviation
    mean_score = np.mean(score_array)
    median_score = np.median(score_array)
    std_dev_score = np.std(score_array)
    print(f"Mean: {mean_score:.2f}")
    print(f"Median: {median_score:.2f}")
    print(f"Standard Deviation: {std_dev_score:.2f}")

    # 3. Find minimum and maximum scores
    min_score = np.min(score_array)
    max_score = np.max(score_array)
    print(f"Minimum: {min_score:.2f}")
    print(f"Maximum: {max_score:.2f}")
```

Tasks:
1. Plot histogram of math scores.
2. Plot bar chart of average scores.
3. Create scatter plot between math and writing scores.


```python
import matplotlib.pyplot as plt

# 1. Plot histogram of math scores
plt.figure(figsize=(10, 6))
plt.hist(df['math_score'], bins=20, edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.grid(axis='y', alpha=0.75)
plt.show()

# 2. Plot bar chart of average scores
average_scores = df[['math_score', 'science_score', 'english_score']].mean()
plt.figure(figsize=(10, 6))
average_scores.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Average Scores Across Subjects')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.ylim(0, 100) # Assuming scores are out of 100
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.75)
plt.show()

# 3. Create scatter plot between math and writing scores (assuming English score as writing score)
plt.figure(figsize=(10, 6))
plt.scatter(df['math_score'], df['english_score'], alpha=0.5)
plt.title('Math Score vs. English Score')
plt.xlabel('Math Score')
plt.ylabel('English Score')
plt.grid(True)
plt.show()
```

Tasks:
1. Group scores by gender.
2. Group scores by parental education.
3. Visualize group-wise average scores using bar charts.


```python
import matplotlib.pyplot as plt

# Score columns for analysis
score_cols = ['math_score', 'science_score', 'english_score', 'overall_score']

# 1. Group scores by gender
gender_grouped_scores = df.groupby('gender')[score_cols].mean()
print("\nAverage Scores by Gender:")
display(gender_grouped_scores)

# 2. Group scores by parental education
parent_education_grouped_scores = df.groupby('parent_education')[score_cols].mean()
print("\nAverage Scores by Parental Education:")
display(parent_education_grouped_scores)

# 3. Visualize group-wise average scores using bar charts

# Bar chart for Average Scores by Gender
gender_grouped_scores.plot(kind='bar', figsize=(12, 7), rot=0)
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.legend(title='Subject')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()

# Bar chart for Average Scores by Parental Education
parent_education_grouped_scores.plot(kind='bar', figsize=(14, 7), rot=45)
plt.title('Average Scores by Parental Education')
plt.xlabel('Parental Education')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.legend(title='Subject')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```
